# Python List Methods – 20 Coding Questions


## Question 1
Create a list of 5 employee names and add a new employee using `append()`.


```python
employees = ["Amit", "Rahul", "Sneha", "Priya", "Kiran"]
employees.append("Arjun")
print(employees)
```

## Question 2
Create a list of account numbers and add three new account numbers using `extend()`.


```python
accounts = [1001, 1002, 1003]
accounts.extend([1004, 1005, 1006])
print(accounts)
```

## Question 3
Insert a new customer at the beginning of a customer list using `insert()`.


```python
customers = ["Rahul", "Priya", "Kiran"]
customers.insert(0, "Amit")
print(customers)
```

## Question 4
Insert a transaction amount at index 3 in a list.


```python
transactions = [1000, 2000, 3000, 4000]
transactions.insert(3, 3500)
print(transactions)
```

## Question 5
Remove the last item from a list using `pop()`.


```python
items = ["Account", "Loan", "Card", "Deposit"]
items.pop()
print(items)
```

## Question 6
Remove the customer "Rahul" from a customer list using `remove()`.


```python
customers = ["Amit", "Rahul", "Priya", "Kiran"]
customers.remove("Rahul")
print(customers)
```

## Question 7
Count how many times the account type "Savings" appears in a list using `count()`.


```python
types = ["Savings", "Current", "Savings", "Salary", "Savings"]
print(types.count("Savings"))
```

## Question 8
Find the index of "Current" in a list of account types using `index()`.


```python
types = ["Savings", "Salary", "Current", "Fixed Deposit"]
print(types.index("Current"))
```

## Question 9
Reverse a list of transaction IDs using `reverse()`.


```python
ids = [101, 102, 103, 104, 105]
ids.reverse()
print(ids)
```

## Question 10
Sort a list of account balances in ascending order using `sort()`.


```python
balances = [50000, 25000, 75000, 10000, 40000]
balances.sort()
print(balances)
```

## Question 11
Sort a list of customer names in descending order using `sort()`.


```python
customers = ["Amit", "Rahul", "Sneha", "Priya", "Kiran"]
customers.sort(reverse=True)
print(customers)
```

## Question 12
Make a copy of a list using `copy()` and verify that modifying the copied list does not affect the original list.


```python
original = [1000, 2000, 3000]
copied = original.copy()
copied.append(4000)
print("Original:", original)
print("Copied:", copied)
```

## Question 13
Clear all elements from a list using `clear()`.


```python
accounts = [1001, 1002, 1003]
accounts.clear()
print(accounts)
```

## Question 14
Append a nested list containing a new customer's details to an existing customer list.


```python
customers = [["Amit", 1001], ["Priya", 1002]]
customers.append(["Rahul", 1003])
print(customers)
```

## Question 15
Extend a list of branch names with another list of new branches.

## Question 16
Remove the first occurrence of the balance `50000` from a list using `remove()`.


```python
balances = [25000, 50000, 75000, 50000]
balances.remove(50000)
print(balances)
```

## Question 17
Pop the element at index 2 from a list and print the removed value.


```python
values = [10, 20, 30, 40, 50]
removed = values.pop(2)
print("Removed value:", removed)
print(values)
```

Count how many times the city "Bangalore" appears in a list.
