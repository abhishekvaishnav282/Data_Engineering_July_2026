## Question 28
Calculate compound interest using `math.pow()`.


```python
import math

principal = 10000
rate = 5
time = 2

amount = principal * math.pow((1 + rate / 100), time)
compound_interest = amount - principal
print(compound_interest)
```

## Question 30
Create a scientific calculator that supports square root, factorial, logarithm, power, trigonometric functions, GCD, and LCM.


```python
import math

# Example values for the calculator
num = 25
n = 5
a = 12
b = 18
angle = 45

print("Square Root:", math.sqrt(num))
print("Factorial:", math.factorial(n))
print("Logarithm:", math.log10(100))
print("Power:", math.pow(2, 3))
print("Sine:", math.sin(math.radians(angle)))
print("Cosine:", math.cos(math.radians(angle)))
print("Tangent:", math.tan(math.radians(angle)))
print("GCD:", math.gcd(a, b))
print("LCM:", math.lcm(a, b))
```
